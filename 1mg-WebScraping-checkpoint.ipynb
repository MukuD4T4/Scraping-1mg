```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
url = 'https://www.1mg.com/categories/fitness-supplements/vitamins-minerals-9?filter=true&pageNumber=1'
response=requests.get(url,headers=headers)
```

```python
soup=BeautifulSoup(response.content,'lxml')
```

```python
new_dict = {
    "Name" : [],
    "pack_size" : [],
    "rating_count" : [],
    "rating" : [],
    "price" : [],
    "mrp" : [],
    }
```

```python
for item in soup.select('.row style__add-padding___1bTbb style__grid-container___1Xvb-'):
    try:
        #print(item)
        new_dict["Name"].append(item.select('.style__pro-title___2QwJy').get_text())
        new_dict["pack_size"].append(item.select('.style__pack-size___2JQG7').get_text())
        new_dict["rating_count"].append(item.select('.CardRatingDetail__ratings-header___2yyQW').get_text())
        new_dict["rating"].append(item.select('.CardRatingDetail__weight-700___27w9q').get_text())
        new_dict["price"].append(item.select('.style__price-tag___cOxYc').get_text())
        new_dict["mrp"].append(item.select('.style__discount-price___25Bya').get_text())
        print("----------------------")
       
    except Exception as e:
        print('Error')
        
```

```python

for i in soup.select('.style__container___1TL2R'):
    new_dict["Name"].append(i.select('.style__pro-title___2QwJy')[0].get_text())
    new_dict["pack_size"].append(i.select('.style__pack-size___2JQG7')[0].get_text())
    try:
        new_dict["rating_count"].append(i.select('.CardRatingDetail__ratings-header___2yyQW')[0].get_text())
    except:
        new_dict["rating_count"].append('No reviews')
    try:
        new_dict["rating"].append(i.select('.CardRatingDetail__weight-700___27w9q')[0].get_text())
    except:
        new_dict["rating"].append('Nan')
    new_dict["mrp"].append(i.select('.style__price-tag___cOxYc')[0].get_text())
    try:
        new_dict["price"].append(i.select('.style__discount-price___25Bya')[0].get_text())
    except:
        new_dict["price"].append('No discounts')
```

```python
final=pd.DataFrame()
for j in range(1,400):
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
    url = 'https://www.1mg.com/categories/fitness-supplements/vitamins-minerals-9?filter=true&pageNumber={}'.format(j)
    response=requests.get(url,headers=headers)
    soup=BeautifulSoup(response.content,'lxml')
    new_dict = {
    "Name" : [],
    "pack_size" : [],
    "rating_count" : [],
    "rating" : [],
    "price" : [],
    "mrp" : [],
    }
    
    for i in soup.select('.style__container___1TL2R'):
        new_dict["Name"].append(i.select('.style__pro-title___2QwJy')[0].get_text())
        new_dict["pack_size"].append(i.select('.style__pack-size___2JQG7')[0].get_text())
        try:
            new_dict["rating_count"].append(i.select('.CardRatingDetail__ratings-header___2yyQW')[0].get_text())
        except:
            new_dict["rating_count"].append('No reviews')
        try:
            new_dict["rating"].append(i.select('.CardRatingDetail__weight-700___27w9q')[0].get_text())
        except:
            new_dict["rating"].append('Nan')
        new_dict["mrp"].append(i.select('.style__price-tag___cOxYc')[0].get_text())
        try:
            new_dict["price"].append(i.select('.style__discount-price___25Bya')[0].get_text())
        except:
            new_dict["price"].append('No discounts')
    df = pd.DataFrame.from_dict(new_dict, orient='index')
    df=df.transpose()
    final=final.append(df)
    

    
```
